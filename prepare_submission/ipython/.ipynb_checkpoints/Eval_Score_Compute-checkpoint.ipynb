```python
import json
import pickle as pkl
from scipy import stats
def load_pkl(path):
    data=pkl.load(open(path,'rb'))
    return data

def load_json(path):
    data=json.load(open(path,'r'))
    return data

import pandas as pd
from collections import defaultdict
```

```python
path=os.path.abspath('../..')
print (path)
p2_data=load_json(os.path.join(path,'data/data_clean/split_data/test.json'))
test_id_to_row={}
for i,row in enumerate(p2_data):
    test_id_to_row[i]=row
```

```python

llm_name='qwen'
mllm_name='qwen'

#llm_name='qwen'
#mllm_name='llava'

llm_name='gemma'
mllm_name='gemma'

#llm_name="gemini-2.0-flash-001"
#mllm_name="gemini-2.0-flash-001"
save_num='12'

pred_file=load_pkl(os.path.join(path,'fc_detailed_results','_'.join([llm_name,mllm_name]),save_num+'.pkl'))
print (len(pred_file))
```

```python
score_file=load_json(os.path.join(path,
                                  'prepare_submission/intermediate_eval_results',
                                  '_'.join([llm_name,mllm_name,str(save_num)])+'.json'))
print (len(score_file))
```

```python
ques_scores=0.
justi_scores=0.
evid_scores=0.
verdict_scores=0.
threshold=0.4

for i,row in enumerate(score_file):
    ques_score=row['ques_score']
    evid_score=row['evid_score']
    justi_score=row['justi_score']
    verdict_score=row['verdict_score']
    gt_verdict=p2_data[i]['label']
    pred_verdict=pred_file[i]['verdict']
    #print (pred_verdict.lower().strip()==gt_verdict.lower(),gt_verdict,pred_verdict)
    

    ques_scores+=ques_score
    evid_scores+=evid_score
    if evid_score>threshold:
        verdict_scores+=verdict_score
        justi_scores+=justi_score

print ('##Question:',ques_scores/len(score_file))
print ('##Evidence:',evid_scores/len(score_file))
print ('##Verdict:',verdict_scores/len(score_file))
print ('##Justification:',justi_scores/len(score_file))
```
